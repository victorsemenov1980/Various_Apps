
# Chart 1: Current Call Open Interest
proceed to the next chart if you don't need this one. Each set of cells relating to separate chart can be run separately with no relation with running or not running other cells


```python
output_notebook() 



plot_width = 400
plot_height = 400

strikes=Calls['Strike'] 
op_int=Calls['Open Int']
s1 = ColumnDataSource(data=dict(x=strikes,y=op_int))
MyHover10 = HoverTool(
	tooltips=[
		( 'Strike', '@x'),
		( 'OI', '@y' ), 
       
	], point_policy="follow_mouse"	
)    
p = figure(title="Current Call OI, select-here chart",tools=["pan,zoom_in, xzoom_in, yzoom_in,zoom_out, xzoom_out, yzoom_out,box_select", MyHover10], x_axis_label='Strike Price', y_axis_label='OI',width=plot_width,height=plot_height)
p.yaxis.formatter.use_scientific = False
MyHover10.formatters.use_scientific = False
p.vbar(source=s1,x='x', width=0.5, bottom=0,
        top='y', color="firebrick")



s2 = ColumnDataSource(data=dict(x=[], y=[]))

fig02 = figure(
    plot_width=400,
    plot_height=400,
    tools=["box_zoom", "wheel_zoom", "reset", "save"],
    title="Watch Here",
)

fig02.vbar(source=s2,x='x', width=0.5, bottom=0,
        top='y', color="firebrick")

# create dynamic table of selected points
columns = [
    TableColumn(field="x", title="Strike"),
    TableColumn(field="y", title="OI"),
]

table = DataTable(
    source=s2,
    columns=columns,
    width=400,
    height=600,
    sortable=True,
    selectable=True,
    editable=True,
)

s1.selected.js_on_change(
    "indices",
    CustomJS(
        args=dict(s1=s1, s2=s2, table=table),
        code="""
        var inds = cb_obj.indices;
        var d1 = s1.data;
        var d2 = s2.data;
        d2['x'] = []
        d2['y'] = []
        for (var i = 0; i < inds.length; i++) {
            d2['x'].push(d1['x'][inds[i]])
            d2['y'].push(d1['y'][inds[i]])
        }
        s2.change.emit();
        table.change.emit();
        IPython.notebook.kernel.execute("foo ="+inds)
       
    """,
    ),
)


layout = grid([p, fig02, table], ncols=3)

output_notebook()
show(layout)
```
